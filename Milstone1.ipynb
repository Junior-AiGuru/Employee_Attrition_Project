```python
# فك التشفير
df["Department"] = df.filter(like="Department_").idxmax(axis=1).str.replace("Department_", "")
df["EducationField"] = df.filter(like="EducationField_").idxmax(axis=1).str.replace("EducationField_", "")
```

```python
# تحويل Attrition من Yes/No إلى 1/0
df["Attrition"] = df["Attrition"].map({"Yes": 1, "No": 0})

# One-Hot Encoding للأعمدة الفئوية (مثال: Department)
df = pd.get_dummies(df, columns=["Department", "EducationField"])
```

```python
from sklearn.preprocessing import StandardScaler

# تطبيع الأعمدة الرقمية (مثال: Age, MonthlyIncome)
scaler = StandardScaler()
df[["Age", "MonthlyIncome"]] = scaler.fit_transform(df[["Age", "MonthlyIncome"]])
```

```python
# تحويل OverTime إلى 0 و1
df["OverTime"] = df["OverTime"].map({"Yes": 1, "No": 0})

# تحويل MonthlyIncome إلى رقمي
df["MonthlyIncome"] = pd.to_numeric(df["MonthlyIncome"], errors="coerce")

# حذف القيم المفقودة
df.dropna(subset=["MonthlyIncome"], inplace=True)

# حساب النسبة
df["OvertimeToIncome"] = df["OverTime"] / df["MonthlyIncome"]

# تصنيف سنوات الخدمة
bins = [0, 2, 5, 10, 40]
labels = ["Junior", "Mid", "Senior", "Veteran"]
df["TenureGroup"] = pd.cut(df["YearsAtCompany"], bins=bins, labels=labels)
```

```python
df.to_csv('cleaned_employee_data.csv', index=False)
```
